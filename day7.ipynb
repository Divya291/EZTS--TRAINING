```python
class node:
    def __init__(self,data):
        self.value=data
        self.left=None
        self.right=None

def preorder(root):
    if root == None:
        return
        
    print(root.value)
    preorder(root.left)
    preorder(root.right)

def inorder(root):
    if root == None:
        return
        
    inorder(root.left)
    print(root.value)
    inorder(root.right)

def postorder(root):
    if root == None:
        return
        
    postorder(root.left)
    postorder(root.right)
    print(root.value)

if __name__=="__main__":
    root = node(1)

    root.left = node(2)
    root.right = node(3)

    root.left.left = node(4)
    root.left.right = node(5)

    root.right.left = node(6)
    root.right.right = node(7)

    print("preorder")
    preorder(root)
    print("Inorder")
    inorder(root)
    print("postorder")
    postorder(root)
```

```python
class node:
    def __init__(self,data):
        self.value=data
        self.left=None
        self.right=None

def height(root):
    if root == None:
        return 0
    LH = height(root.left)
    RH = height(root.right)
    return max(LH,RH)+1
    

if __name__=="__main__":
    root = node(1)
    root.left = node(2)
    root.right = node(5)
    root.left.left = node(3)
    root.left.right = node(4)
    root.right.right = node(6)
    root.left.right.left = node(9)
    root.right.right.left = node(7)
    root.right.right.right = node(8)
    root.left.right.left.right = node(10)
    root.right.right.left.right = node(11)
    root.left.right.left.right.left = node(14)
    root.right.right.left.right.left = node(12)
    root.right.right.left.right.left.right = node(13)

    print(height(root))

```

```python
class node:
    def __init__(self,data):
        self.value=data
        self.left=None
        self.right=None
class node_data:
    def __init__(self,node,HKey):
        self.node=node
        self.Hkey=HKey
def Top_View(root):
    temp=node_data(root,0)
    Key_dic={}
    Q=[temp]
    Q.append(None)
    while len(Q)!=0:
        cur=Q.pop(0)
        if cur==None:
            if len(Q)==0:
                break
            else:
                Q.append(None)
        else:
            if cur.Hkey not in Key_dic.keys():
                Key_dic[cur.Hkey]=cur.node.value
            if cur.node.left!=None:
                temp=node_data(cur.node.left,cur.Hkey-1)
                Q.append(temp)
            if cur.node.right!=None:
                temp=node_data(cur.node.right,cur.Hkey+1)
                Q.append(temp)
    for i in sorted(Key_dic):
        print(Key_dic[i],end=' ')
    print(Key_dic)
if __name__=="__main__":
    root=node(1)
    root.left=node(2)
    root.right=node(5)
    root.left.left=node(3)
    root.left.right=node(4)
    root.right.right=node(6)
    root.left.right.left=node(9)
    root.right.right.left=node(7)
    root.right.right.right=node(8)
    root.left.right.left.right=node(10)
    root.right.right.left.right=node(11)
    root.left.right.left.right.left=node(14)
    root.right.right.left.right.left=node(12)
    root.right.right.left.right.left.right=node(13)
    
    Top_View(root)
```

```python
class node:
    def __init__(self,data):
        self.value=data
        self.left=None
        self.right=None
class node_data:
    def __init__(self,node,HKey):
        self.node=node
        self.Hkey=HKey
def Bottom_View(root):
    temp=node_data(root,0)
    Key_dic={}
    Q=[temp]
    Q.append(None)
    while len(Q)!=0:
        cur=Q.pop(0)
        if cur==None:
            if len(Q)==0:
                break
            else:
                Q.append(None)
        else:
            
            Key_dic[cur.Hkey]=cur.node.value
            if cur.node.left!=None:
                temp=node_data(cur.node.left,cur.Hkey-1)
                Q.append(temp)
            if cur.node.right!=None:
                temp=node_data(cur.node.right,cur.Hkey+1)
                Q.append(temp)
    for i in sorted(Key_dic):
        print(Key_dic[i],end=' ')
    print(Key_dic)
if __name__=="__main__":
    root=node(1)
    root.left=node(2)
    root.right=node(5)
    root.left.left=node(3)
    root.left.right=node(4)
    root.right.right=node(6)
    root.left.right.left=node(9)
    root.right.right.left=node(7)
    root.right.right.right=node(8)
    root.left.right.left.right=node(10)
    root.right.right.left.right=node(11)
    root.left.right.left.right.left=node(14)
    root.right.right.left.right.left=node(12)
    root.right.right.left.right.left.right=node(13)
    
    Bottom_View(root)
```
